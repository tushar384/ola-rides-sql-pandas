# 1. What are the busiest hours for bookings?
How to Use This Insight:

If you notice a peak at a certain hour (e.g., 8:00 AM):
Focus marketing efforts during that hour (e.g., send promotions before 8:00 AM).

Ensure maximum staff availability to handle demand during peak hours.

Consider time-based offers to maximize revenue during quieter hours (e.g., discounts from 2:00 PM to 4:00 PM).

This graph is about identifying the most profitable time slots of the day for better business decisions, not the result of working 8 hours directly.

```python
import plotly.express as px
import pandas as pd



# Extract hour from the Time column
data['Hour'] = pd.to_datetime(data['Time'], format='%H:%M:%S').dt.hour

# Group by hour and sum the Booking_Value to get total revenue
hourly_revenue = data.groupby('Hour')['Booking_Value'].sum().reset_index(name='Revenue')

# Find the peak hour (highest revenue)
peak_hour = hourly_revenue.loc[hourly_revenue['Revenue'].idxmax(), 'Hour']
peak_revenue = hourly_revenue['Revenue'].max()

# Create the question and answer to show in the graph title
title = f"Hourly Revenue Performance \nA: {peak_hour}:00 - ₹{peak_revenue:,.2f}"

# Plot the hourly revenue
fig = px.bar(hourly_revenue, x='Hour', y='Revenue',
             title=title,
             labels={'Hour': 'Hour of the Day', 'Revenue': 'Total Revenue (₹)'},
             color='Revenue', color_continuous_scale='Viridis')

fig.show()
```

```python
# Extract hour from the Time column
data['Hour'] = pd.to_datetime(data['Time'], format='%H:%M:%S').dt.hour

# Group by hour and count bookings
hourly_bookings = data.groupby('Hour').size().reset_index(name='Bookings')

# Plot
fig = px.bar(hourly_bookings, x='Hour', y='Bookings', title='Busiest Hours for Bookings',
             labels={'Hour': 'Hour of the Day', 'Bookings': 'Number of Bookings'})
fig.show()

# Answer
busiest_hour = hourly_bookings.loc[hourly_bookings['Bookings'].idxmax(), 'Hour']
print(f"The busiest hour for bookings is {busiest_hour}:00.")
```

# 2.Which vehicle type generates the highest revenue?

```python
# Group by Vehicle_Type and sum Booking_Value
revenue_by_vehicle = data.groupby('Vehicle_Type')['Booking_Value'].sum().reset_index()

# Plot
fig = px.pie(revenue_by_vehicle, values='Booking_Value', names='Vehicle_Type',
             title='Revenue by Vehicle Type')
fig.show()

# Answer
highest_revenue_vehicle = revenue_by_vehicle.loc[revenue_by_vehicle['Booking_Value'].idxmax(), 'Vehicle_Type']
print(f"The vehicle type that generates the highest revenue is {highest_revenue_vehicle}.")
```

# 3.What is the average ride distance for different vehicle types?

```python
# Group by Vehicle_Type and calculate mean Ride_Distance
avg_distance = data.groupby('Vehicle_Type')['Ride_Distance'].mean().reset_index()

# Plot
fig = px.bar(avg_distance, x='Vehicle_Type', y='Ride_Distance',
             title='Average Ride Distance by Vehicle Type',
             labels={'Ride_Distance': 'Average Distance (km)'})
fig.show()

# Answer
avg_distance_sorted = avg_distance.sort_values(by='Ride_Distance', ascending=False)
for _, row in avg_distance_sorted.iterrows():
    print(f"Average distance for {row['Vehicle_Type']} is {row['Ride_Distance']:.2f} km.")
```

# 4. Which locations have the highest number of incomplete rides, and why?

```python
# Filter incomplete rides and group by Pickup_Location
incomplete_rides = data[data['Incomplete_Rides'] == 'Yes']
location_incomplete = incomplete_rides.groupby('Pickup_Location').size().reset_index(name='Count')

# Plot
fig = px.bar(location_incomplete, x='Pickup_Location', y='Count',
             title='Incomplete Rides by Pickup Location',
             labels={'Pickup_Location': 'Location', 'Count': 'Number of Rides'})
fig.show()

# Answer
top_location = location_incomplete.loc[location_incomplete['Count'].idxmax(), 'Pickup_Location']
reason = incomplete_rides['Incomplete_Rides_Reason'].mode().values[0]
print(f"The location with the highest number of incomplete rides is {top_location}, "
      f"with the most common reason being '{reason}'.")
```

# 6.Which location has the highest ride frequency?

```python
ride_frequency = data.groupby('Pickup_Location').size().reset_index(name='Frequency')

fig = px.bar(ride_frequency, x='Pickup_Location', y='Frequency',
             title='Ride Frequency by Location',
             labels={'Pickup_Location': 'Location', 'Frequency': 'Number of Rides'})
fig.show()

# Answer
top_location = ride_frequency.loc[ride_frequency['Frequency'].idxmax(), 'Pickup_Location']
print(f"The location with the highest ride frequency is {top_location}.")
```

# 7.1. Most Common Pickup Locations (Bar Chart)

```python
import plotly.express as px

# Count occurrences of each pickup location
pickup_counts = data['Pickup_Location'].value_counts().reset_index()
pickup_counts.columns = ['Pickup_Location', 'Count']

# Plot the top 5 pickup locations
fig = px.bar(pickup_counts.head(5), x='Pickup_Location', y='Count',
             title='Top 5 Pickup Locations', text='Count')

# Show counts inside the bars
fig.update_traces(textposition='inside')
fig.update_layout(xaxis_title='Pickup Location', yaxis_title='Total Pickups')
fig.show()
```

# 7.2. Top Pickup-Drop Location Pairs (Bar Chart)

```python
# Combine pickup and drop locations into pairs
data['Location_Pair'] = data['Pickup_Location'] + " → " + data['Drop_Location']
pair_counts = data['Location_Pair'].value_counts().reset_index()
pair_counts.columns = ['Location_Pair', 'Count']

# Plot the top 5 location pairs
fig = px.bar(pair_counts.head(5), x='Location_Pair', y='Count',
             title='Top 5 Pickup-Drop Location Pairs', text='Count')

# Show counts inside the bars
fig.update_traces(textposition='inside')
fig.update_layout(xaxis_title='Location Pair', yaxis_title='Total Rides')
fig.show()
```

# 8. Which day of the week generates the most revenue?

```python
data['Weekday'] = data['Date'].dt.day_name()
weekday_revenue = data.groupby('Weekday')['Booking_Value'].sum().reset_index()

fig = px.bar(weekday_revenue, x='Weekday', y='Booking_Value',
             title='Revenue by Day of the Week',
             labels={'Booking_Value': 'Revenue'})
fig.show()

# Answer
top_day = weekday_revenue.loc[weekday_revenue['Booking_Value'].idxmax(), 'Weekday']
print(f"The day with the highest revenue is {top_day}.")
```

# 9. What is the revenue contribution of each customer segment?

```python
import plotly.express as px
import pandas as pd

# Calculate descriptive statistics
mean = data['Booking_Value'].mean()
std = data['Booking_Value'].std()

# Define ranges
low_upper = mean - std
medium_upper = mean
high_upper = mean + std

print(f"This is my Low Upper {low_upper}")
print(f"This is my Medium Upper {medium_upper}")
print(f"This is my High Upper {high_upper}")

# Add a column to categorize Booking_Value
bins = [0, low_upper, medium_upper, high_upper, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']
data['Customer_Segment'] = pd.cut(data['Booking_Value'], bins=bins, labels=labels)

# Plot the histogram
fig = px.histogram(
    data,
    x='Booking_Value',
    nbins=20,
    color='Customer_Segment',
    title='Distribution of Booking Values with Segments',
    labels={'Booking_Value': 'Booking Value', 'Customer_Segment': 'Segment'}
)

# Add vertical lines to represent thresholds
fig.add_vline(x=low_upper, line_dash="dash", line_color="blue", annotation_text="Low Upper Limit")
fig.add_vline(x=medium_upper, line_dash="dash", line_color="green", annotation_text="Medium Upper Limit")
fig.add_vline(x=high_upper, line_dash="dash", line_color="orange", annotation_text="High Upper Limit")
fig.add_vline(x=mean, line_dash="dot", line_color="red", annotation_text="Mean")

# Show the plot
fig.show()
```
